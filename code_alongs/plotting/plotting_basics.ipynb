# Plotting basics

#### https://github.com/mwaskom/seaborn-data

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/titanic.csv")
df.info()
```

```python
import matplotlib.pyplot as plt
# we use different plt methods!
plt.hist(df["age"], bins=50)
plt.xlabel("age")
plt.ylabel("count")
plt.title("Histogram of ages in titanic dataset")
```

#### OOP approach in matplotlib

```python
fig, ax = plt.subplots(1) # fig contains many ax objects
ax.hist(df["age"], bins=50) # with semicolon we mark that we are done with this fig
ax.set(title="Histogram of ages in titanic dataset", ylabel=("count"),xlabel=("age") )

# check: https://matplotlib.org/stable/gallery/showcase/anatomy.html
```

**bar chart in matplotlib**

```python
df.columns
```

```python
classes = df["class"].value_counts().reset_index() # we change series back to df
classes
```

```python
fig, ax = plt.subplots(1)
ax.bar(x = classes["class"], height = classes["count"])
ax.set(xlabel = "class", ylabel = "count")
```

```python
cohorts = ["DE23", "DE24", "AI23", "JAVA24"]
number_students = [ 29, 38, 28, 20]

fig, ax = plt.subplots(1)
ax.bar(x = cohorts, height = number_students)
```
